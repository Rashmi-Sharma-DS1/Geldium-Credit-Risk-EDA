#### **Missing Values Summary**

A small number of values are missing across three key financial columns.

| Column | Missing Count | Imputation Strategy (Industry Best Practice) | Rationale |
| :--- | :--- | :--- | :--- |
| $\mathbf{Income}$ | $\mathbf{7}$ | **Median Imputation** | Protects against outlier distortion, as income can be highly skewed. |
| $\mathbf{Credit\_Score}$ | $\mathbf{1}$ | **Median Imputation** | A single missing value; median is a stable, non-parametric measure. |
| $\mathbf{Loan\_Balance}$ | $\mathbf{6}$ | **Median Imputation** | Loan balances can be skewed by large loans, making the median a more robust estimate. |

```python
# CODE CELL 2: Missing Data Handling (Median Imputation - Step 2 Action)
df_cleaned = df.copy()
```

```python
# Identify columns for median imputation
imputation_cols = ['Income', 'Credit_Score', 'Loan_Balance']

print("--- Missing Values Before Imputation ---")
print(df_cleaned[imputation_cols].isnull().sum())
```

```python
# Calculate and apply median imputation
for col in imputation_cols:
    median_val = df_cleaned[col].median()
    df_cleaned[col].fillna(median_val, inplace=True)
    print(f"Imputing '{col}' with median value: {median_val:.2f}")
```

```python
print("\n--- Missing Values After Imputation (Verification) ---")
print(df_cleaned[imputation_cols].isnull().sum())
```

```python
# CODE CELL 4: Visualize Monthly Payment Failure Trends

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

target = 'Delinquent_Account'

# --- 1. Calculate Monthly Failure Rate ---
payment_cols = ['Month_1', 'Month_2', 'Month_3', 'Month_4', 'Month_5', 'Month_6']

# Convert 'Late' and 'Missed' to 1 (Failure), 'On-time' to 0 (Success)
df_payment = df_cleaned[payment_cols].replace({'Late': 1, 'Missed': 1, 'On-time': 0})

# Calculate the failure rate (mean of the binary column) for each month
monthly_failure_rate = (df_payment.sum() / len(df_payment)).sort_values(ascending=False)
monthly_failure_rate = (monthly_failure_rate * 100).round(2) # Convert to percentage

print("\nMonthly Payment Failure Rate Trend (Percentage):")
print(monthly_failure_rate)

# --- 2. Visualize the Trend (Bar Chart) ---
plt.figure(figsize=(10, 4))
# Use a color palette to highlight the highest risk months (e.g., using a single accent color)
sns.barplot(x=monthly_failure_rate.index, y=monthly_failure_rate.values,palette="coolwarm")

# Add the percentage values on top of the bars
for index, value in enumerate(monthly_failure_rate.values):
    plt.text(index, value + 0.5, f'{value}%', ha='center', va='bottom', fontsize=10)

plt.title('Monthly Payment Failure Rate (Late/Missed)', fontsize=14)
plt.ylabel('Failure Rate (%)', fontsize=12)
plt.xlabel('Payment Month', fontsize=12)
plt.ylim(0, 100) # Set Y-limit for better comparison
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.figtext(0.95, 0.01, 'Rashmi sharma .', fontsize=8, color='#000000', style='italic',
             verticalalignment='bottom', horizontalalignment='right')
# Display the plot inline
plt.show()
```

#### **B. Top 3 Financial Risk Factors**

The most significant factors associated with current account delinquency ($\mathbf{Delinquent\_Account}=1$) relate to historical credit behavior and financial strain.

| Rank | Risk Factor | Delinquent (Avg.) | Non-Delinquent (Avg.) | Impact on Delinquency |
| :--- | :--- | :--- | :--- | :--- |
| **1** | $\mathbf{Missed\_Payments}$ (Count) | $\approx \mathbf{3.53}$ | $\approx \mathbf{2.94}$ | **Highest Predictor:** Delinquent customers have a much higher history of missed payments. |
| **2** | $\mathbf{Credit\_Utilization}$ (Ratio) | $\approx \mathbf{0.505}$ | $\approx \mathbf{0.457}$ | **High Strain:** Indicates greater use of available credit, suggesting liquidity issues. |
| **3** | $\mathbf{Credit\_Score}$ | $\approx \mathbf{551.4}$ | $\approx \mathbf{597.5}$ | **Poor Credit History:** A substantially lower credit score implies a weaker overall financial standing. |

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- STEP 1: Load and Clean Data (Re-run Imputation for safety) ---
imputation_cols = ['Income', 'Credit_Score', 'Loan_Balance']

# Apply Median Imputation
for col in imputation_cols:
    median_val = df_cleaned[col].median()
    df_cleaned[col].fillna(median_val, inplace=True)

# --- STEP 2: Data Aggregation for Plotting ---
target = 'Delinquent_Account'
risk_indicators = ['Missed_Payments', 'Credit_Utilization', 'Credit_Score', 'Income']

# Calculate the mean of key risk indicators grouped by delinquency status
risk_summary = df_cleaned.groupby(target)[risk_indicators].mean().T.reset_index()
risk_summary.columns = ['Risk_Factor', 'Non_Delinquent_0', 'Delinquent_1']

# Reshape the data for plotting (long format)
df_plot = risk_summary.melt(id_vars='Risk_Factor', 
                            var_name='Delinquency_Status', 
                            value_name='Average_Value')

# Rename status for better labels
df_plot['Delinquency_Status'] = df_plot['Delinquency_Status'].replace({
    'Non_Delinquent_0': 'Non-Delinquent',
    'Delinquent_1': 'Delinquent'
})

# --- Visualization: Grouped Bar Charts using Subplots ---
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

# Define titles, order, and labels
factor_titles = {
    'Missed_Payments': '1. Average Missed Payments (Count)',
    'Credit_Utilization': '2. Average Credit Utilization (Ratio)',
    'Credit_Score': '3. Average Credit Score',
    'Income': '4. Average Income ($)'
}
plot_order = ['Missed_Payments', 'Credit_Utilization', 'Credit_Score', 'Income']
y_labels = {
    'Missed_Payments': 'Count',
    'Credit_Utilization': 'Ratio',
    'Credit_Score': 'Score',
    'Income': 'Average Value ($)'
}

# Define the custom coolwarm palette for binary comparison
# Blue (Cool) for Non-Delinquent, Red (Warm) for Delinquent
coolwarm_palette = {
    'Non-Delinquent': sns.color_palette("coolwarm")[0], 
    'Delinquent': sns.color_palette("coolwarm")[-1]
}


for i, factor in enumerate(plot_order):
    data = df_plot[df_plot['Risk_Factor'] == factor]
    
    # Plot using sns.barplot with the requested coolwarm palette
    sns.barplot(x='Delinquency_Status', y='Average_Value', data=data, ax=axes[i], 
                palette=coolwarm_palette, 
                order=['Non-Delinquent', 'Delinquent'])
    
    # Customize the plot
    axes[i].set_title(factor_titles[factor], fontsize=14)
    axes[i].set_xlabel('Customer Group', fontsize=12)
    axes[i].set_ylabel(y_labels[factor], fontsize=12)
    
    # Annotate bars with exact values
    for p in axes[i].patches:
        height = p.get_height()
        offset = height * 0.05 if height > 10 else 0.05
        axes[i].text(p.get_x() + p.get_width() / 2.,
                     height + offset,
                     f'{height:,.2f}',
                     ha='center', fontsize=11)

plt.tight_layout(pad=3.0)
fig.suptitle('B. Comparative Analysis of Key Financial Risk Factors', fontsize=16, y=1.02)

# Remember to use plt.show() in your live notebook
# plt.show()
```

```python
# CODE CELL 1: VISUALIZE MISSED PAYMENTS (Part B - Risk Factor 1)
# Filter data for the specific factor
data = df_plot[df_plot['Risk_Factor'] == 'Missed_Payments']

# Define the coolwarm palette
coolwarm_palette = {'Non-Delinquent': sns.color_palette("coolwarm")[0], 'Delinquent': sns.color_palette("coolwarm")[-1]}

plt.figure(figsize=(6, 5))
sns.barplot(x='Delinquency_Status', y='Average_Value', data=data, 
                 palette=coolwarm_palette, 
                 order=['Non-Delinquent', 'Delinquent'])

plt.title('Average Missed Payments (Count) by Customer Group', fontsize=14)
plt.xlabel('Customer Group', fontsize=12)
plt.ylabel('Average Count', fontsize=12)
plt.ylim(0, 4) 
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Annotate bars
for p in plt.gca().patches: # Use plt.gca() to get the current axes
    plt.gca().text(p.get_x() + p.get_width() / 2., p.get_height() + 0.1, 
            f'{p.get_height():,.2f}', ha='center', fontsize=11)

plt.show()
```
